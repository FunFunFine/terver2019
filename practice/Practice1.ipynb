### Основные формулы
$$P = \frac{m}{n} = \frac{\textrm{число благоприятных исходов}}{\textrm{число всех исходов}}$$

$$P = \frac{S_{A}}{S}= \frac{\textrm{площадь благоприятных исходов}}{\textrm{вся площадь}}$$


Число способов выбрать k объектов из n различных объектов.
$$C_{n}^{k} = \frac{n!}{k!(n-k)!}$$

Число перестановок
$$P_n = n!$$

### Задачи на пару

**Задача 1.** На полке в случайном порядке расставлено 40 книг, среди которых находится трехтомник А. С. Пушкина. Найти вероятность того, что эти тома стоят в порядке возрастания слева направо (но не обязательно рядом).

**Задача 2.** Партия продукции состоит из десяти изделий, среди которых два изделия дефектные. Какова вероятность того, что из пяти отобранных
наугад и проверенных изделий:
* ровно одно изделие дефектное;
* ровно два изделия дефектные;
* хотя бы одно изделие дефектное?

**Задача 3.** 2 лыжника условились о встрече между 10 и 11 часами утра у подно-
жия горы (склона), причем договорились ждать друг друга не более
10 минут, чтобы не замерзнуть. Считая, что момент прихода на встречу каждым выбирается наудачу в пределах указанного часа, найдите вероятность того, что встреча состоится.

**Задача 4** (Задача Бюффона.) Плоскость разграфлена параллельными прямыми, отстоящими друг от друга на расстоянии $2a$. На плоскости наудачу, бросается игла длины $2l (l < a)$. Найти вероятность того, что игла пересечет какую-нибудь прямую.

## ДЗ

**Задача 1.** В программе к экзамену по теории вероятностей 75 вопросов. Студент знает 50 из них. В билете 3 вопроса. Найдите вероятность того, что студент знает хотя бы два вопроса из вытянутого им билета.

**Задача 2.** (Парадокс дней рождений) Найти вероятность того, что в классе из 23 человек, не менее двух учеников родились в один день.

**Задача** У театральной кассы стоят в очереди 2n человек. Среди них n человек имеют лишь банкноты по 1000 рублей, а остальные — только банкноты по 500 рублей. Билет стоит 500 рублей. Каждый покупатель приобретает по одному билету. В начальный момент в кассе нет денег. Чему равна вероятность того, что никто не будет ждать сдачу?

**Задача** На плоскости, замощённой одинаковыми прямоугольниками со сторонами 10 и 20 (прямоугольники примыкают сторонами), рисуют случайную окружность радиуса 4. Найдите вероятность того, что окружность имеет общие точки ровно с тремя прямоугольниками.

**Задача** Отрезок длины a случайным образом разделен на 3 части. Найти вероятность того, что длины хотя бы двоих частей не меньше, чем $a/5$.

### Основные формулы
$$P = \frac{m}{n} = \frac{\textrm{число благоприятных исходов}}{\textrm{число всех исходов}}$$

$$P = \frac{S_{A}}{S}= \frac{\textrm{площадь благоприятных исходов}}{\textrm{вся площадь}}$$


Число способов выбрать $k$ объектов из $n$ различных объектов.
$$C_{n}^{k} = \frac{n!}{k!(n-k)!}$$

Число перестановок
$$P_n = n!$$

## ДЗ

**Задача 1.** В программе к экзамену по теории вероятностей 75 вопросов. Студент знает 50 из них. В билете 3 вопроса. Найдите вероятность того, что студент знает хотя бы два вопроса из вытянутого им билета.

Пусть билеты занумерованы:

$$ \Omega = \{(a_i,a_j,a_k) \mid a_i,a_j,a_k \in [1;75] \,\,\text{without repetitions}\} $$
тогда
$ \|\Omega\|  = C_{75}^3 $;

Пусть студент знает билеты от $1$ до $50$. Тогда необходимое событие выглядит так:
$$ A = \{(a_i,a_j,a_k) \mid a_i,a_j \in [1;50], \, a_k \in [51;75]\,\,\text{without repetitions}\} $$
и
$$ \| A\| = C_{50}^{2} \cdot  C_{25}^{1} = C_{50}^{2} \cdot 25 = 49 \cdot 25^2 $$



**Задача 2.** (Парадокс дней рождений) Найти вероятность того, что в классе из 23 человек, не менее двух учеников родились в один день.

Пусть элементарное событие -- $23$-ка чисел, каждое обозначает порядковый номер дня рождения $i$-го по порядку ученика:
$$ \Omega = \{(a_1, a_2,...a_i,..., a_{23}) \mid a_i \in [1;365] \}$$

$$\|\Omega\| = A_{365}^{23} $$.

Интересующее нас событие выглядит как $(a_i)_{i=1}^{23}$, где совпадают $2$ или больше $a_i$.
Пусть дни рождения совпадают у всех, таких вариантов дней 365 штук.
Если же дни рождения совпадут у  22 человек и у 1 отличный от других, то можем выбрать таких дней $365 \cdot 365$ штук.

Нехитрыми рассуждениями получаем, что количество вариантов равно $\sum_{n=1}^{23}365^n$.
Таким образом получаем:

$$ P = \frac{\sum_{n=1}^{23}365^n}{A_{365}^{23} } $$


**Задача 3.** У театральной кассы стоят в очереди 2n человек. Среди них n человек имеют лишь банкноты по 1000 рублей, а остальные — только банкноты по 500 рублей. Билет стоит 500 рублей. Каждый покупатель приобретает по одному билету. В начальный момент в кассе нет денег. Чему равна вероятность того, что никто не будет ждать сдачу?

Пусть элементарное событие &mdash; $(a_1, a_2, \ldots, a_{2n})$, где
$$ a_i = \begin{cases}
    ')' & \text{если $i$-ый человек имеет 1000 рублей} \\
    '(' & \text{в ином случае}
  \end{cases} $$
  
Тогда чтобы каждый "тысячник" не ждал сдачу, перед ним должен быть в очереди хоть один "пятисотник". Представив людей в виде скобок сводим задачу к поиску количества возможных правильных скобочных последовательностей &mdash; открывающий предшествует закрывающему.  
Интернет говорит, что такое количество равно $C_n$ &mdash; $n$-ому числу Каталана. 
В свою очередь, всего последовательностей равно $2^{2n}$.

Таким образом получаем
$$ P = \frac{C_n}{2^{2n}} = \frac{{2n \choose n} - {2n \choose n - 1}}{2^{2n}} $$

**Гроб 4.** Из чисел $\{1, 2, \ldots, N\}$ случайно выбирается число $a$. Найти вероятность $p_N$ того, что:

* (b) число $a$ не делится ни на какое из чисел $a_1, a_2, \ldots, a_k$, где числа $a_i$ -- натуральные и попарно взаимно простые.

Найти $\lim_{N\rightarrow\infty} p_N$ в случаях (a) и (b).

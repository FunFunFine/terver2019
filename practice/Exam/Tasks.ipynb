- Равновероятно выбирается строка $r = s$ или $r = t$.
- Выбираются ровно $15$ различных позиций (каждое сочетание $15$ элементов равновероятно).
- В выбранных позициях в строке r биты инвертируются.
- Полученную строку берут в качестве результата.

Валя утверждает, что вероятность того, что в строке $x$, в позиции $y$ символ был инвертирован $p = 0.15$. Прав ли он?
